## Data Transformation

#### Converting Object to Datetime Datatype

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
```

#### Converting Categorical Columns Datatype AS Category Datatype

```python
categorical_cols = [
'Ship Mode',
'Segment',
'Country',
'City',
'State',
'Region',
'Category',
'Sub-Category'
]

for col in categorical_cols:
    data[col] = data[col].astype('category')
```

#### Converting Keys Columns Datatype AS String Datatype

```python
id_cols = [
'Order ID',
'Customer ID',
'Product ID',
'Postal Code'
]

for col in id_cols:
    data[col] = data[col].astype('string')
```

#### Renaming column's names to snake case

```python
data.columns = (
    data.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("-", "_")
)
```

#### Data Transformation

Four things were done to get the dataset analysis-ready: date parsing, categorical encoding, identifier typing, and column renaming.

#### Transformations Applied

**Datetime Conversion**  
`Order Date` and `Ship Date` were converted from object to datetime. Without this, any time-based calculation — trends, delivery gaps, seasonality — won't work correctly.

**Categorical Conversion**  
Ship Mode, Segment, Country, City, State, Region, Category, and Sub-Category were converted to category dtype. These columns have a small number of repeating values, so storing them as categories cuts memory usage and speeds up groupby operations noticeably.

**String Conversion for Identifiers**  
Order ID, Customer ID, Product ID, and Postal Code were cast to string. They're labels, not numbers — keeping them as objects or integers risks unintended arithmetic and inconsistent joins downstream.

**Column Renaming**  
All column names were renamed to snake_case. Consistent naming removes the friction of quoting columns with spaces in Pandas, SQL, and Power BI.

#### Result

The dataset is now typed correctly, lighter in memory, and consistent in naming — ready to load into any analysis or visualization layer without further prep.


## Feature Engineering

#### Creating Date Features Like Year & Month

```python
data['order_year'] = data['order_date'].dt.year
data['order_month'] = data['order_date'].dt.month
```

#### Creating Shipping Days Column

```python
data['shipping_days'] = (data['ship_date'] - data['order_date']).dt.days
```

#### Creating Profit Margin & Percentage Columns

```python
data['profit_margin'] = data['profit'] / data['sales']
```

```python
data['profit_margin_pct'] = (data['profit'] / data['sales']) * 100
```

#### Feature Engineering

Four new columns were added to make the data more directly useful for business analysis. None of these existed in the raw dataset — they're all derived from existing fields.

#### Features Created

**Date Features**  
Year and Month were extracted from `order_date`. Having these as standalone columns makes time-based SQL and Power BI straightforward — no need to parse dates repeatedly during analysis.

**Shipping Duration**  
`shipping_days` was created by subtracting `order_date` from `ship_date`. This gives a concrete measure of delivery speed per order, useful for spotting logistics bottlenecks by region, ship mode, or time period.

**Profit Metrics**  
Two profitability columns were added: `profit_margin` (profit relative to sales) and `profit_percentage`. These make it easier to compare performance across products, categories, and regions without doing the calculation every time in a query or visual.

#### Result

The dataset now has the time, logistics, and profitability dimensions pre-built. Downstream analysis can work directly with these columns instead of re-deriving them at every step.

#### Spark DataFrame Creation & Delta Table Storage

At this point the data is clean, typed, and validated. The last step is moving it out of Pandas processing and into a format the rest of the pipeline can actually use at scale.

#### Process Breakdown

**Pandas to Spark**  
The cleaned Pandas DataFrame was converted to a Spark DataFrame using `spark.createDataFrame(data)`. This hands off the data to distributed processing — necessary once the pipeline moves beyond a single notebook or dataset size grows.

**Data Preview**  
`display(spark_df)` was used to do a quick sanity check on the schema and first few records before writing anything to storage. Small step, but worth doing before an overwrite.

**Delta Table Storage**  
The Spark DataFrame was written using `write.format("delta").mode("overwrite").saveAsTable("silver_sales_cleaned")`. Delta Lake handles the storage layer here — it gives ACID transactions, version history (time travel if something breaks), and consistent reads even during writes. Overwrite mode replaces the existing table cleanly without stacking duplicates.

**Silver Layer Placement**  
The table is saved as `silver_sales_cleaned`, which places it in the Silver layer of the medallion architecture: Bronze holds raw ingested data, Silver holds cleaned and transformed data, and Gold holds business-ready aggregations. This notebook covers the Bronze → Silver step.

#### Result

The transformed dataset is now stored as a managed Delta table, queryable from SQL, consumable by downstream notebooks, and ready to feed into the Gold layer
